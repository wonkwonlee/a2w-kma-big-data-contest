# ASOS데이터에 행정동 데이터 합치기

```python
# 토양도에서 행정동 데이터 가져오기
dist = merge_s_h_t[['umd', 'sgg', 'umd2']].copy()
dist
```

```python
# sgg의 앞에 2글자 추출하여 loc에 저장
dist['loc'] = dist['sgg'].str[0:2]
dist
```

```python
# 특정 지역 loc변수 수정
for i in range(len(dist['loc'])):
  if dist['loc'][i] == '울릉':
    dist['loc'][i] = '울릉도'
  elif dist['loc'][i] == '김해':
    dist['loc'][i] = '김해시'
  elif dist['loc'][i] == '양산':
    dist['loc'][i] = '양산시' 
  elif dist['loc'][i] == '의령':
    dist['loc'][i] = '의령군'
  elif dist['loc'][i] == '함양':
    dist['loc'][i] = '함양군'
  elif dist['loc'][i] == '청송':
    dist['loc'][i] = '청송군'
  elif dist['loc'][i] == '경주':
    dist['loc'][i] = '경주시'
  # 창원의 경우 북창원과 창원 구분이 필요함
  elif dist['loc'][i] == '창원시 마산합포구':
    dist['loc'][i] = '창원'
  elif dist['loc'][i] == '창원시 마산회원구':
    dist['loc'][i] = '북창원'
  elif dist['loc'][i] == '창원시 성산구':
    dist['loc'][i] = '북창원'
  elif dist['loc'][i] == '창원시 진해구':
    dist['loc'][i] = '창원'
  elif dist['loc'][i] == '창원시 의창구':
    dist['loc'][i] = '북창원'
```

```python
# ASOS 날씨 데이터 기준 sites
sites = ['115', '130', '136', '137', '138', '143', '152', '155', '159', '253', '255', '257', '263', '264', '271', '272', '273', '162', '192', '276', '277', '278', '279', '281', '283', '284', '285', '288', '289', '294', '295']

sites_index = ['울릉도', '울진', '안동', '상주', '포항', '대구', '울산', '창원', '부산', '김해시', '북창원', '양산시', '의령군', '함양군', '봉화', '영주', '문경', '통영', '진주', '청송군', '영덕', '의성', '구미', '영천', '경주시', '거창', '합천', '밀양', '산청', '거제', '남해']

dic = {key:value for key, value in zip(sites_index, sites)}

dist['stnId'] = 0
missing_area = []

# loc값에 맞는 stnId 매칭
for i in range(len(dist['loc'])):
  if dist['loc'][i] in sites_index:
    dist['stnId'][i] = dic[dist['loc'][i]]
  else:
    missing_area.append(dist['loc'][i])
```

```python
# 매칭 확인
dist
```

```python
# 가장 가까운 기상대
closest_area = {'김천': '구미', '예천': '문경', '사천': '진주', '청도': '밀양', '성주': '구미', 
                '고성': '구미', '고령': '합천', '함안': '창원', '하동': '진주', '창녕': '밀양',
                '경산': '영천', '칠곡': '구미', '군위': '의성'}

for i in range(len(dist['loc'])):
  # 영양을 제외한 다른 구역은 가장 가까운 기상대 위치로 날씨 측정
  if dist['loc'][i] in closest_area:
    # closest_area 딕셔너리에 value 값을 key 값 삼아서 그에 해당하는 dic 딕셔너리 sites 코드 value 값 리턴
    curr = closest_area[dist['loc'][i]]
    dist['stnId'][i] = dic[curr]
    # dist['stnId'][i] = dic[closest_area[dist['loc'][i]]
  # 영양의 경우 읍면동 단위에 따라 서로 다른 기상대 위치로 날씨 측정
  if dist['loc'][i] == '영양':
    if dist['umd'][i] == '석보면' or dist['umd'][i] == '영양읍' or dist['umd'][i] == '입암면':
      dist['stnId'][i] = dic['청송군']
    elif dist['umd'][i] == '일월면' or dist['umd'][i] == '수비면':
      dist['stnId'][i] = dic['울진']
    elif dist['umd'][i] == '청기면':
      dist['stnId'][i] = dic['안동']
  if dist['loc'][i] == '창원':
    if dist['sgg'][i] == '창원시 의창구' or dist['sgg'][i] == '창원시 마산회원구' or dist['sgg'][i] == '창원시 성산구':
      dist['stnId'][i] = dic['북창원']
```

```python
# ASOS 데이터와 전처리한 산사태 발생 이력 데이터 결합
# 기준은 tm (날짜), stnId (측정소 ID)

asos_merge = pd.merge(asos, dist, how='outer', on=['stnId'])
asos_merge
```

```python
# umd2 기준으로 토양도와 임상도와 날씨 데이터 병합
merge_s_h_a = pd.merge(asos_merge, merge_s_t, how='inner',on='umd2')
merge_s_h_a
```

```python
# 새로운 산사태 발생 이력 데이터프레임 생성

# 산사태 발생 이력 데이터 불러오기
landslide_CNT = pd.read_csv('landslide_history.csv', encoding='cp949')

# 산사태 발생 이력에 시와 읍면동 기준으로 umd2 생성
landslide_CNT["umd2"] = landslide_CNT["sgg"] + " " + landslide_CNT["umd"]

# 날짜형 데이터
landslide_CNT["date"] = landslide_CNT["date"].astype("str")
landslide_CNT["date"] = pd.to_datetime(landslide_CNT["date"])
landslide_CNT = landslide_CNT.drop(["sd", 'umd', "sgg", "sum_cnt", "sum_hpa"], axis=1)
landslide_CNT.rename(columns={'date':'tm'}, inplace=True)
landslide_CNT["landslide"] = 1

landslide_CNT
```

```python
# 날씨, 임상도, 토양도가 합쳐진 데이터와 산사태 발생 이력 데이터 병합 (umd2와 날짜 기준)

merge_final = pd.merge(merge_s_h_a, landslide_CNT, how='inner',on=['tm','umd2'])
merge_final
```
