```python
import pandas as pd
# 소수점 2자리까지 표시
pd.options.display.float_format = '{:.2f}'.format

df = pd.read_csv('soil_land.csv')
df
```

```python
# 범주형 변수 분석 함수
def check_categorical(check_data, check_data_vc):
  feature = check_data_vc.columns[0]
  col1 =  feature, 'land0'
  col2 =  feature, 'land1'

  df1 = pd.DataFrame(check_data.columns[:-1], columns = [feature])


  values = check_data_vc[check_data_vc['landslide']==1].value_counts().tolist()
  keys = check_data_vc[check_data_vc['landslide']==1].value_counts().index.tolist()
  indexs = [keys_index[0] for keys_index in keys]

  A1 = dict(zip(indexs, values)) # 1

  df2 = pd.DataFrame(list(A1.items()), columns = col2)


  values2 = check_data_vc[check_data_vc['landslide']==0].value_counts().tolist()
  keys2 = check_data_vc[check_data_vc['landslide']==0].value_counts().index.tolist()
  indexs2 = [keys_index[0] for keys_index in keys2]

  A2 = dict(zip(indexs2, values2)) # 1

  df3 = pd.DataFrame(list(A2.items()), columns = col1)

  df_merge = pd.merge(df1,df2, how='outer',on=feature)
  df_merge = pd.merge(df_merge,df3, how='outer',on=feature)
  df_merge = df_merge.fillna(0)

  df_merge['ratio'] = df_merge['land1'] / (df_merge['land1'] + df_merge['land0'])

  display(df_merge)

  return df_merge
```

```python
# 각 index별로 나눗셈을 보는 과정

values = check_data8_vc[check_data8_vc['landslide']==1].value_counts().tolist()
keys = check_data8_vc[check_data8_vc['landslide']==1].value_counts().index.tolist()

indexs = [keys_index[0] for keys_index in keys]
print(indexs)

A1 = dict(zip(indexs, values)) # 1

values2 = check_data8_vc[check_data8_vc['landslide']==0].value_counts().tolist()
keys2 = check_data8_vc[check_data8_vc['landslide']==0].value_counts().index.tolist()

indexs2 = [keys_index[0] for keys_index in keys2]
print(indexs2)

A2 = dict(zip(indexs2, values2)) # 0

B1 = dict(sorted(A1.items())) # 1
B2 = dict(sorted(A2.items())) # 0
print(B1, B2)

from collections import Counter
C = dict(Counter(B1) + Counter(B2)) # 1과 0의 전체 합

print(C)

final_result = {x:float(B1[x])/C[x] for x in C}
print(final_result)
```
