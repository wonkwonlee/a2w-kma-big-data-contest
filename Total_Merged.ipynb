```python
import pandas as pd
import csv
```

```python
# 수정할 ASOS 데이터 불러오기
ASOS_final = pd.read_csv('datasets/ASOS_final.csv')
ASOS_final.head()
```

```python
# 합칠 hg 데이터 불러오기
hg = pd.read_csv('datasets/sum_hg_all.csv')
hg.head()
```

```python
# tree
tree = pd.read_csv('datasets/tree_sum_final.csv', encoding = 'cp949')
tree.head()
```

```python
# 분류
'''
김천 - 구미
영양 - 청송 (석보, 영양) / 울진 (수비, 일월)
예천 - 문경
사천 - 진주
청도 - 밀양
성주 - 김천
고성 - 구미
고령 - 합천
함안 - 창원
하동 - 진주 (광양)
창녕 - 밀양
'''

# ASOS + hg 데이트 합치기
# input : 날짜 + 지역
input_date = '2011-05-31' # 날짜 단위
input_ssg = '포항' # 시 단위
input_umd = '기계면' # 읍면동 단위

def specific_data(ssg, umd, date):
    # 지역 조건과 날짜 조건을 만족하는 ASOS 데이터
    ASOS_final_specific = ASOS_final[ASOS_final['stnNm'].str.contains(ssg) & ASOS_final['tm'].str.contains(date)]
    hg_final_specific = hg[hg['umd2'].str.contains(umd)]
    hg_final_specific['stnNm'] = hg_final_specific['umd2'].str[0:2]
    tree_final_specific = tree[tree['umd2'].str.contains(umd)]
    tree_final_specific['stnNm'] = tree_final_specific['umd2'].str[0:2]
    temp_merged_specific = pd.merge(hg_final_specific, ASOS_final_specific, on = 'stnNm')
    merged_specific = pd.merge(tree_final_specific, temp_merged_specific, on = 'stnNm')
    return merged_specific

print(specific_data(input_ssg, input_umd, input_date))
```
